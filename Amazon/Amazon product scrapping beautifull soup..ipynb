```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
with open('Amazon.com _ laptop.html', 'r', encoding='utf-8') as file:
    html = file.read()
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
title_tag = soup.title

title = title_tag.text

print("Page Title:", title)
```

```python
links = soup.find_all('a')

for link in links:
    print("Text:", link.text)
    print("URL:", link.get('href'))
```

```python
divs = soup.find_all('div', class_='puisg-col-inner')
divs
```

```python
valid_products = []

for div in divs:
    product_name = div.find('h2', class_='a-size-mini a-spacing-none a-color-base s-line-clamp-2')
    product_price = div.find('span', class_='a-offscreen')
    product_rating_tag = div.find('i', class_='a-icon-star-small')

    if product_name and product_price and product_rating_tag:
        product_name_text = product_name.a.text.strip()
        product_price_text = product_price.text.strip()
        product_rating_text = product_rating_tag.find('span', class_='a-icon-alt').text.strip()

        valid_products.append({
            "Product Name": product_name_text,
            "Product Price": product_price_text,
            "Product Rating": product_rating_text
        })
```

```python
for product in valid_products:
    print("Product Name:", product["Product Name"])
    print("Product Price:", product["Product Price"])
    print("Product Rating:", product["Product Rating"])
    print("---")
```

```python
products_data = []

for product in valid_products:
    products_data.append({
        "Product Name": product["Product Name"],
        "Product Price": product["Product Price"],
        "Product Rating": product["Product Rating"]
    })

df = pd.DataFrame(products_data)
df
```

```python
df.to_csv('Product_info.csv', index=False)
```
